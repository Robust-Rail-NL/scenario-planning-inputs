```python
import os, json

filename = "scenario_solver_2.json"
path = os.path.join(os.path.dirname(os.path.abspath("__file__")), "..", filename)
scenario = json.load(open(path))
arrivals = [(f"Train{train['id']}", int(train['arrival']), f"({','.join([u['trainUnit']['type']['displayName'] + str(u['trainUnit']['type']['carriages']) for u in train['members']])})") for train in scenario["in"]["trains"]]
arrivals.sort(key=lambda x: x[1])
departures = [(int(train['departure']),f"({','.join([u['type']['displayName'] + str(u['type']['carriages']) for u in train['trainUnits']])})") for train in scenario["out"]["trainRequests"]]
departures.sort(key=lambda x: x[0])
print(arrivals)
print(departures)
```
